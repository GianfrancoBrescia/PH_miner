### *1. Obiettivo di ricerca*
L'obiettivo principale di questa ricerca √® di capire quali fattori portano al successo del lancio di un prodotto su Product Hunt (PH), dove il successo √® definito se un prodotto √® presente tra i pi√π famosi del giorno corrente sulla homepage di PH. I prodotti popolari sono votati dai membri della comunit√†.
I prodotti sono scoperti da un "hunter" su PH. Gli hunter sono membri della comunit√† che "scavano" continuamente su Internet per cercare sempre nuovi prodotti. A volte l'hunter √® anche uno dei maker del prodotto che vuole lanciare e pubblicizzare un nuovo prodotto.

### *2. Fattori di scuccesso di un prodotto*
Come fattori di successo sono stati considerati i seguenti:
* **Reputation (Reputazione)**
* **Presentation quality (Qualit√† della presentazione)**
* **Time (Tempo)**
* **Affect (or Sentiment, Sentimento)**
* **Linguistic (Linguistica)**

Di seguito verr√† spiegato il significato di ogni fattore:
* **Reputation:** questo fattore riguarda sia gli hunter che i maker
* **Presentation quality:** questo fattore √® molto importante. In PH vengono utilizzate, ad esempio, tagline brevi e accattivanti, gif animate, emoji e video di Youtube. Anche la lunghezza della descrizione del testo potrebbe avere un effetto positivo per il lancio del prodotto.
* **Time:** questo fattore riguarda il momento in cui il prodotto √® stato "cacciato" o lanciato. Il refresh della homepage di PH viene fatto giornalmente alle ore 00:01 (Pacific Ocean Time) e il momento migliore per postare √® entro le ore 9:00 in modo che ci sia tempo sufficiente durante l'intera giornata affinch√® le persone possano scoprire e discutere riguardo il prodotto lanciato. Anche sui giorni della settimana in cui bisogna postare c'√® una raccomandazione, infatti √® consigliabile lanciare un prodotto il Luned√¨ o il Marted√¨.
* **Affect:** l'orientamento positivo, negativo o neutro trasmesso dalla descrizione testuale dei commenti dei maker o dalla descrizione dei post potrebbe influenzare oppure no il numero di voti ricevuti.
* **Linguistic:** questo fattore riguarda il topic modeling (modellazione degli argomenti). Nel machine learning (apprendimento automatico) e nel natural language processing (elaborazione del linguaggio naturale), il topic model √® un tipo di modello statistico per scoprire i topic (argomenti) astratti che si presentano in una raccolta di documenti. Il topic modeling √® uno strumento di estrazione del testo frequentemente utilizzato per la scoperta di strutture semantiche nascoste nel testo.

````python
import emoji
import regex

def __extract_emoji(text):
    emoji_list = []
    data = regex.findall(r'\X', str(text))
    for word in data:
        if any(character in emoji.UNICODE_EMOJI for character in word):
            emoji_list.append(word)
    return emoji_list
````

````python
def get_gcd(a, b):
    """ The GCD (greatest common divisor) is the highest number that evenly divides both width and height. """
    return a if b == 0 else get_gcd(b, a % b)


def calculate_aspect_ratio(width, height):
    temp = 0
    if width == height:
        return 1, 1
    if width < height:
        temp = width
        width = height
        height = temp

    divisor = get_gcd(width, height)
    x = int(width / divisor) if not temp else int(height / divisor)
    y = int(height / divisor) if not temp else int(width / divisor)
    return x, y

for p in _posts:
    index_media = 0
    while index_media < len(_media): 
        # check if the current post_id is equal to the post_id of the current media
        if p.id == _media[index_media][0]: 
            # calculate the image size passing its width and height 
            roughly, ratio = calculate_aspect_ratio(_media[index_media][3], _media[index_media][4]) 
            if (roughly == 2) and (ratio == 1):
                # append to the list the image url
                tweetable_images = tweetable_images + [_media[index_media][5]] 
        index_media = index_media + 1
    if tweetable_images:
        entry = ['Yes']
    else:
        entry = ['No']
````

&nbsp;&nbsp;&nbsp;Metrica che indica la versione del prodotto. Operando sul link di ogni post questa metrica √® stata calcolata utilizzando un'espressione regolare basata sulla &nbsp;&nbsp;&nbsp;parola-chiave `post/app-versione` ed eseguibile importando il package `regex`

````python
import regex

def get_keyword_based_heuristics(_keyword):
    return regex.compile(_keyword)


def __extract_version(url):
    url = url + ' '
    version_number = '1.0'
    keyword = get_keyword_based_heuristics(r'-+v+([\d-]+)+\s|-+([\d-]+)+\s')
    _version = keyword.search(url)
    if _version:
        v = _version.group()
        v = v.replace('-', '.')
        if v[1] == 'v':
            if v.count('.') >= 2:
                version_number = v[2:]
            else:
                version_number = v[2:len(v)-1] + '.0'
        else:
            if v.count('.') >= 2:
                version_number = v[1:]
            else:
                version_number = v[1:len(v)-1] + '.0'
    return version_number

for p in _posts:
    version = __extract_version(p.discussion_url)
    entry = [version]
````

**Linguistic.** Si √® operato sul fattore linguistic nel seguente modo: <br>
√® stato usato l'algoritmo **LDA (Latent Dirichlet Allocation)** fornito dalla libreria di python ```Scikit-Learn```, la pi√π popolare libreria di machine learning (in questo progetto √® stata usata la versione 0.21.2). <br>
Per elaborare i testi sono state utilizzate le librerie per espressioni regolari ```regex```, ```gensim```, ```spacy```; per la visualizzazione grafica √® stata utilizzata la libreria ```matplotlib```, infine per la manipolazione e la visualizzazione dei dati in forma tabellare sono state utilizzate le librerie ```numpy``` e ```pandas```.

````python
import os
import numpy as np
import pandas as pd
import regex
import spacy
import gensim

# Sklearn
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import GridSearchCV
from pprint import pprint

# Plotting tools
import matplotlib.pyplot as plt
````

````python
# Import dataset
csv_directory = os.getcwd()[:-8] + 'dataset\\'
csv_file_name = 'features_not_discretized.csv'
csv_path = os.path.join(csv_directory, csv_file_name)
df = pd.read_csv(csv_path, delimiter=';', usecols=['post_name', 'post_description', 'post_tagline'])
df.head(15)
````

Per ogni post sono stati rimossi dalla *tagline* e dalla *description* email, caratteri di nuova linea (\n) e spazi extra. 

````python
# Remove email and new line characters
def remove_email_and_new_line_chars(text):
    _data = text.values.tolist()  # Convert to list
    _data = [regex.sub(r'\S*@\S*\s?', '', str(sent)) for sent in _data]  # Remove Emails
    _data = [regex.sub(r'\s+', ' ', str(sent)) for sent in _data]  # Remove new line characters
    _data = [regex.sub("\'", "", sent) for sent in _data]  # Remove distracting single quotes
    return _data

data = remove_email_and_new_line_chars(df.post_tagline)
pprint(data[:1])
print('\n')
data = remove_email_and_new_line_chars(df.post_description)
pprint(data[:1])
````

Inoltre, ogni frase √® stata tokenizzata in una lista di parole eliminando totalmente punteggiatura e caratteri non necessari. Per fare questo √® stata utilizzata la funzione ```simple_preprocess()``` fornita dalla libreria ```gensim```.

````python
# Tokenize and Clean-up text
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data_words = list(sent_to_words(data))
print(data_words[:1])
````

Successivamente √® stata effettuata la lemmatizzazione, cio√® un processo che riporta le parole alla propria radice. Questo processo serve a ridurre il numero totale delle parole univoche presenti nel dizionario e come risultato il numero di colonne nella matrice documento-parola sar√† pi√π denso con meno colonne.

````python
# Lemmatization
def lemmatization(texts, allowed_postags, nlp):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent))
        texts_out.append(" ".join([token.lemma_ if token.lemma_ not in ['-PRON-'] else '' for token in doc if token.pos_ in allowed_postags]))
    return texts_out

# Initialize spacy 'en' model, keeping only tagger component (for efficiency)
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])  

# Do lemmatization keeping only Noun, Adj, Verb, Adverb
data_lemmatized = lemmatization(data_words, ['NOUN', 'ADJ', 'VERB', 'ADV'], nlp)  
                                                                                      
print(data_lemmatized[:2])
````

A questo punto √® stata creata la matrice documento-parola utilizzando la classe ```CountVectorizer```, in quanto l'algoritmo LDA per topic modeling richiede una matrice documento-parola come input principale. Innanzitutto √® stata inizializzata la classe CountVectorizer per considerare le parole che sono presenti almeno 10 volte, per rimuovere le stop-word e per convertire in minuscolo tutte le parole. Successivamente √® stata applicata la funzione ```fit_transform()``` per creare la matrice. <br>
Poich√® la maggior parte delle celle contiene il valore 0, il risultato sar√† sottoforma di una matrice sparsa per risparmiare memoria.

````python
# Create the Document-Word matrix
vectorizer = CountVectorizer(analyzer='word',
                             min_df=10,                        # minimum required occurences of a word
                             stop_words='english',             # remove english stop words
                             lowercase=True,                   # convert all words to lowercase
                             token_pattern='[a-zA-Z0-9]{3,}',  # number of chars in a word > 3
                             )
data_vectorized = vectorizer.fit_transform(data_lemmatized)
````

La sparsit√† non √® altro che la percentuale di punti dati diversi da 0 nella matrice documento-parola (```data_vectorized```) e per sapere qual √® questa percentuale viene eseguito il seguente codice:

````python
# Check the Sparsicity
data_dense = data_vectorized.todense()  # Materialize the sparse data
print("Sparsicity: ", ((data_dense > 0).sum() / data_dense.size) * 100, "%")  # Compute Sparsicity = Percentage of Non-Zero cells
````

A questo punto √® stato inizializzato un modello LDA settando il numero di topic a 10. Inoltre, √® stata richiamata la funzione ```fit_transform()``` per costruire il modello LDA.

````python
# Build LDA model with sklearn
lda_model = LatentDirichletAllocation(n_components=10,            # Number of topics
                                      max_iter=10,                # Max learning iterations
                                      learning_method='online',
                                      random_state=100,           # Random state
                                      batch_size=128,             # number of documents in each learning iter
                                      evaluate_every=-1,          # compute perplexity every n iters, default: Don't
                                      n_jobs=-1,                  # Use all available CPUs
                                      )
lda_output = lda_model.fit_transform(data_vectorized)
print(lda_model)  # Model attributes
````

Successivamente sono state controllate le prestazioni del modello LDA sulla base di perplessit√† e verosimiglianza. Un modello con una pi√π alta verosimiglianza e una minore perplessit√† √® considerato buono.

````python
# Diagnose model performance with perplexity and log-likelihood
print("Log Likelihood: ", lda_model.score(data_vectorized))  # Log Likelyhood: Higher the better
print("Perplexity: ", lda_model.perplexity(data_vectorized))   # Perplexity: Lower the better. Perplexity = exp(-1. * log-likelihood per word)
pprint(lda_model.get_params())  # See model parameters
````

In seguito √® stato cercato il miglior modello LDA prendendo in considerazione due importanti parametri di ottimizzazione per i modelli LDA: ```n_components``` (numero di topic) e ```learning_decay``` (controlla il tasso di apprendimento).

````python
# GridSearch the best LDA model
search_params = {'n_components': [3], 'learning_decay': [.5, .7, .9]}  # Define Search Param
model = GridSearchCV(lda_model, param_grid=search_params, n_jobs=1, iid=True, cv=3, error_score='raise')  # Init Grid Search Class
model.fit(data_vectorized)  # Do the Grid Search
````

````python
# Find the best topic model and its parameters
best_lda_model = model.best_estimator_  # Best Model
print("Best Model's Params: ", model.best_params_)  # Model Parameters
print("Best Log Likelihood Score: ", model.best_score_)  # Log Likelihood Score
print("Model Perplexity: ", best_lda_model.perplexity(data_vectorized))  # Perplexity
````

Inoltre, √® stato visualizzato il topic dominante in ciascun documento. Per stabilire qual √® il topic dominante in un documento bisogna vedere quale topic ha il maggior contributo a quel documento e assegnarlo. <br>
Nella tabella seguente, si sono segnati in verde tutti i topic principali in un documento ed evidenziato in giallo il topic pi√π dominante.

````python
# Create Document - Topic Matrix to see the dominant topic in each document
def create_document_topic_matrix(blm, doc_word_matrix, data):
    lda_output = blm.transform(doc_word_matrix)  
    topicnames = ["Topic" + str(i) for i in range(blm.n_components)]  # column names
    docnames = ["Doc" + str(i) for i in range(len(data))]  # index names
    df_doc_topic = pd.DataFrame(np.round(lda_output, 2), columns=topicnames, index=docnames)  # Make the pandas dataframe
    return df_doc_topic

# Styling
def color_green(val):
    color = 'green' if val > .1 else 'black'
    return 'color: {col}'.format(col=color)

def make_bold(val):
    weight = 700 if val > .1 else 400
    return 'font-weight: {weight}'.format(weight=weight)

# Apply Style
df_document_topic = create_document_topic_matrix(best_lda_model, data_vectorized, data)
n_documents = 15
df_document_topics = df_document_topic.head(n_documents).style.applymap(color_green).applymap(make_bold).highlight_max(color='yellow', axis=1)
df_document_topics
````

Infine per ogni topic sono state prese le prime 15 keyword con il punteggio pi√π alto utilizzando la funzione ```show_topics()```.

````python
def show_topics(vectorizer, best_lda_model, n_words):
    keywords = np.array(vectorizer.get_feature_names())
    topic_keywords = []
    for topic_weights in best_lda_model.components_:
        found = 0
        key_words = []
        top_keyword_locs = (-topic_weights).argsort()[:n_words]
        for i in range(0, len(keywords.take(top_keyword_locs))):
            if keywords.take(top_keyword_locs)[i] == 'nan':
                found = 1
                top_keyword_locs = (-topic_weights).argsort()[:n_words+1]
                key_words.append(keywords.take(top_keyword_locs)[i+1])
            else:
                if found == 0:
                    key_words.append(keywords.take(top_keyword_locs)[i])
                else:
                    key_words.append(keywords.take(top_keyword_locs)[i+1])
        topic_keywords.append(key_words)
    return topic_keywords

topic_keywords = show_topics(vectorizer, best_lda_model, 15)  # Show top n keywords for each topic. In this case n is equal to 15
df_topic_keywords = pd.DataFrame(topic_keywords)  # Topic- Keywords Dataframe
df_topic_keywords.columns = ['Word ' + str(i) for i in range(df_topic_keywords.shape[1])]
df_topic_keywords.index = ['Topic ' + str(i) for i in range(df_topic_keywords.shape[0])]
df_topic_keywords
````
